```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
routed_counts = []
with open("/root/lm-evaluation-harness/results/mixtral-8x7b-reroute-threshold-0.9-Opt-K2-GToken-WikiText-Debug-PenalizeNewTokens-Alpha5/log.txt", "r") as f:
    first_time = True
    for line in f:
        if "Total rerouted tokens" in line:
            rerouted_tokens = [int(x.strip()) for x in line.split("Total rerouted tokens")[1].split("[")[1].split("]")[0].split(",")]
            if first_time:
                first_time = False
                continue
            routed_counts.extend(rerouted_tokens)

```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 3))
sns.histplot(routed_counts, ax=ax, bins=100)
ax.set_title("Rerouted Tokens Distribution (Alpha = 5)")
```

```python
fig.savefig("rerouted_tokens_distribution_alpha1.pdf", bbox_inches='tight')
```
